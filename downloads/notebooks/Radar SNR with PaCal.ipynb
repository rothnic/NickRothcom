```python
P_t = pc.BetaDistr(alpha=1.5, beta=5) * 1000
G = pc.BetaDistr(alpha=1.5, beta=5) * 40
lmda = pc.BetaDistr(alpha=1.5, beta=5) * 8e9
sigma = pc.BetaDistr(alpha=1.5, beta=5) * 5
R = pc.BetaDistr(alpha=1.5, beta=5) * 1000
k = 1.38e-23
T_s = 950
B_n = 1e6
L = 6
```

```python
snr = (P_t * (G**2) * (lmda**2) * sigma)/(((4*pi)**3) * (R**4) * k * T_s * B_n * L)
```

```python
snr_samp = snr.rand(n=10000)
```

```python
sns.distplot(snr_samp)
```
