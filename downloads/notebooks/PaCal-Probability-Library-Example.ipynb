```python
import pacal as pc
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
x = pc.BetaDistr(alpha=2, beta=5)
y = pc.BetaDistr(alpha=10, beta=2)
```

```python
z = x * y
z.plot()
```

```python
zcdf = z.get_piecewise_cdf_interp()
zcdf.plot()
```

```python
samps = z.rand_invcdf(n=1000)
```

```python
sns.kdeplot(samps)
```
